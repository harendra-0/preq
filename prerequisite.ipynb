```python
#Q3

import time

a = list()
b = list()
c = list()
n = int(input())
maX = 0
median = 0
for i in range(n):
    em = random.randint(1, 100)
    a.append(em)
    b.append(em)
    c.append(em)
  

start = time.time_ns()
if n%2 != 0:
    for i in range(n//2):
        for j in range(len(a)):
            if(maX < a[j]):
                maX = a[j]
        a.remove(maX)
        #print(max)
        maX = 0
    for j in range(len(a)):
            if(median < a[j]):
                median = a[j]
    print("Median: ", median)
            
else:
    maX = 0
    for i in range((n//2)-1):
        for j in range(len(a)):
            if(maX < a[j]):
                maX = a[j]
        a.remove(maX)
        maX = 0
    me = 0
    mew = 0
    for j in range(len(a)):
        if(me < a[j]):
            me = a[j]
    a.remove(me)
    for j in range(len(a)):
        if(mew < a[j]):
            mew = a[j]
    a.remove(mew)
    print("Median: ", (me+mew)/2)
    
end = time.time_ns()
print("Time taken: ", (end-start)/10**6)

#better method 
start = time.time_ns()
b.sort()
if n%2 != 0:
    print("Median using sorting: ", b[n//2])
else:
    print("Median using sorting: ", (b[n//2]+b[n//2-1])/2)
end = time.time_ns()
print("Time taken: ", (end-start)/10**6)

#using numpy.median func
start = time.time_ns()
median = np.median(c)
print("Median using numpy: ", median)
end = time.time_ns()
print("Time taken: ", (end-start)/10**6)
```
